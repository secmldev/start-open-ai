```python
model = "davinci"
```

```python
question = "what is the capital of France?"
```

```python
response = openai.Completion.create(
    engine = model,
    prompt=(f"Question: {question}\n"
            "Answer:"
            ),
    max_tokens=100,
    n=1,
    stop=None,
    temperature=0.5
)

answer = response.choices[0].text.split('\n')[0]
print(answer)
```
